# Imports

```python
import pandas as pd
import numpy as np
from sklearn.model_selection  import train_test_split
```

```python
Fraud_Data=pd.read_csv("fraud_detect.csv")
```

```python
#responsible variable analysis

class_name={0:'not found',1:'Fraud'}
print(Fraud_Data.Class.value_counts().rename(index=class_name))
```

```python
#output
y=Fraud_Data["Class"]
```

```python
X=Fraud_Data.loc[:,Fraud_Data.columns!='Class']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=(1/3),random_state=42,stratify=y)
```

# LOGISTIC REGREESSION

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
logisreg=LogisticRegression(max_iter=1000)
```

```python
logisreg.fit(X_train,y_train)
```

```python
y_pred=logisreg.predict(X_test)
acc_logisreg=round(accuracy_score(y_test,y_pred)*100,2)
```

```python
acc_logisreg
```

# GAUSSIAN NAIVE BYES

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gaussianNB=GaussianNB()
gaussianNB.fit(X_train,y_train)
y_pred=gaussianNB.predict(X_test)
acc_gaussian=round(accuracy_score(y_test,y_pred)*100,2)
```

```python
acc_gaussian
```

# RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
acc_rf = round(accuracy_score(y_test, y_pred) * 100, 2)
print("Accuracy of Random Forest : ", acc_rf)
```

# Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decisionTree=DecisionTreeClassifier()
decisionTree.fit(X_train,y_train)
y_pred=decisionTree.predict(X_test)
acc_decisionTree=round(accuracy_score(y_test,y_pred)*100,2)
```

```python
acc_decisionTree
```

```python
best_fit=max(acc_decisionTree,acc_rf,acc_gaussian,acc_logisreg)
print("Best fit model  ")
```

# k nearest neighbour classifier


```python

from sklearn.neighbors import KNeighborsClassifier
```

```python
kneighbour=KNeighborsClassifier()
kneighbour.fit(X_train,y_train)

y_pred=kneighbour.predict(X_test)
acc_kneighbour=round(accuracy_score(y_test,y_pred)*100,2)
```

```python
acc_kneighbour
```

```python
models = pd.DataFrame({'Model': ['Logistic Regression', 'Naive Bayes', 'Decision Tree', 
'Random Forest', 'K - Nearest Neighbors'],'Score': [acc_logisreg, acc_gaussian, acc_decisionTree, acc_rf, acc_kneighbour]})
```

```python
models.sort_values(by='Score',ascending=False)
```
